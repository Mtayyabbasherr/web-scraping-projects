```python
from bs4 import BeautifulSoup 
import requests
```

```python
url='https://en.wikipedia.org/wiki/List_of_largest_companies_in_Pakistan'
page=requests.get(url)
soup=BeautifulSoup(page.text,'html')
```

```python
print(soup)
```

```python
soup.find_all('table')
```

```python
soup.find_all('table',class_='wikitable sortable')
```

```python
table=soup.find_all('table')[1]
```

```python
print(table)
```

```python
name_ind=table.find_all('th')
```

```python
name_ind
```

```python
name_ind_title=[title.text.strip()for title in  name_ind]
print(name_ind_title)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(columns=name_ind_title)
df
```

```python
column_data=table.find_all('tr')
```

```python
for row in column_data [1 :]:
    row_data =row.find_all("td")
    ind_row_data=[data.text.strip() for data in row_data]
    ##print(ind_row_data)
    length=len(df)
    df.loc[length]=ind_row_data
    
```
